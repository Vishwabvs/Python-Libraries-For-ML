#### Matplotlib vc Matlab

Working with Matplotlib in object oriented method.

In this we create objects of figure and then manipulate them

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,1,1])          #[start_x,start_y,width_x,width_y]  
#Even though we change first two values in the list chnages wont be noticed but changes will come when we plot two plots together

axes.plot(x,y,'r')
#plt.xlabel('X Axis') This also works fyn but we have to use one convention
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('Random plot')
plt.show()
```

Multiple plots in one figure

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])   
ax2 = fig.add_axes([0.1,0.6,0.4,0.3])

ax1.plot(x, y, 'r*-')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Random plot Outside')

ax2.plot(x2, y2, 'go-')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Random plot Inside')

plt.show()
```

#### Already we know abt plt.subplot(), now we are going to learn plt.subplots()

```python
fig, ax = plt.subplots(1,2)
#brute-force way
ax[0].plot(x, y, 'b')
ax[1].plot(x, y, 'g')
```

```python
fig,ax = plt.subplots(1,2)
#for loop

for axes in ax:
    axes.plot(x, y)
```

```python
fig,ax = plt.subplots(1,2)
#for loop
col = ['r','g']
Y = [y, y2]
X = [x, x2]
mark = ['*-','o:']
for i,axes in enumerate(ax):
    axes.plot(X[i], Y[i], col[i]+mark[i])
fig.tight_layout()
```

How to change figsize, aspect ratio and dpi

```python
fig,ax = plt.subplots(figsize = (10,4), dpi = 100)

ax.plot(x,y,'r',label = 'y*x')
ax.plot(x2, y2, 'b', label = 'x2^2')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Random plot')
fig.legend(loc = 10)
fig.savefig('Random_plot')    #To save this figure, we can right click and save or we have another method
```

#### How to plot graphs on log scale

```python
fig, ax = plt.subplots(1, 2, figsize = (10,4))

ax[0].plot(x, y, x2, y2)

ax[1].plot(x, np.exp(x))   # Inorder to do log scaling for y-axis do following
ax[1].set_yscale('log')

fig.tight_layout()
```

Set and Reset ticks( markings on x and y axis )

```python
fig, ax = plt.subplots(figsize = (10,5))
ax.plot(x2, y2, lw = 2)                   #lw is line width
ax.set_xticks([2,4,6,8,10])
ax.set_xticklabels(['a', 'b', 'c', 'd', 'e'], fontsize = 19)   # to set characters
ax.set_yticks([0,50,100])

plt.show()
```

Here we arent getting those tick values in scientific notation. How to do?

```python
fig, ax = plt.subplots()
ax.plot(x2,y2)
ax.set_title('Scientific Notation')


from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText= True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter)

plt.show()
```
